```python
# shows the data
foo = pd.read_csv("car_price.csv")
print(foo)
```

```python
# plots the data
fig = plt.figure(figsize=(10,10))

plt.figure(1)
plt.scatter(foo["price"], foo["compressionratio"])
plt.show()
```

```python
# categorizes the data into groups, to make it easier to train
grouped_price = []
for i in range(0,len(foo['price'])):
    if (foo['compressionratio'][i] <= 13):
        grouped_price.append(-1)
    else:
        grouped_price.append(1)
```

```python
# preceptron class
class perceptron:
    def __init__(self, eta, n_iter, inp_size):
        self.eta = eta
        self.n_iter = n_iter
        self.weights = np.random.normal(loc = 0.0, scale = 0.01, size = inp_size)
    def activation_fun(self, z):
        if (z >= 0):
            return 1
        return -1
    
    def fit(self, inp, exp):
        self.progress = []
        for j in range(0,self.n_iter):
            self.errors = []
            out = self.predict(inp)
                
            for i in range(0,len(inp)):
                self.weights[i] += self.eta*(exp[i]-out[i])*inp[i]
                self.errors.append(exp[i]-out[i])
            self.progress.append(sum(self.errors)/len(self.errors))
        
    def predict(self, inp):
        out = []
        for i in range(0,len(inp)):
            out.append(self.activation_fun(inp[i]* self.weights[i]))
        return out
```

```python
# initialize the perceptron
p1 = perceptron(0.01,50, len(foo["price"]))  
```

```python
# graph the predictions before training
prediction = p1.predict(foo["price"])
a_price = []
a_cr = []
b_price = []
b_cr = []

for i in range(0,len(foo["price"])):
    if (prediction[i] == -1):
        a_price.append(foo["price"][i])
        a_cr.append(foo["compressionratio"][i])
    if (prediction[i] == 1):
        b_price.append(foo["price"][i])
        b_cr.append(foo["compressionratio"][i])

        
plt.scatter(a_price, a_cr, label = '-1', color = 'red', marker = 'x')
plt.scatter(b_price, b_cr, label = '1', color = 'blue', marker = 'o')
plt.xlabel("Price")
plt.ylabel("Compression Ratio")
plt.legend(loc='upper right')
plt.show()
```

```python
# train it on the dataset
p1.fit(foo["price"], grouped_price)
```

```python
# show the progression of trianing by plotting average error through each epoch
prediction = p1.predict(foo["price"])
plt.plot(p1.progress)
plt.title("progress of the perceptron over epochs")
plt.xlabel("epochs")
plt.ylabel("error")
plt.show()


# graph the data after training
prediction = p1.predict(foo["price"])
a_price = []
a_cr = []
b_price = []
b_cr = []

for i in range(0,len(foo["price"])):
    if (prediction[i] == -1):
        a_price.append(foo["price"][i])
        a_cr.append(foo["compressionratio"][i])
    if (prediction[i] == 1):
        b_price.append(foo["price"][i])
        b_cr.append(foo["compressionratio"][i])

        
plt.scatter(a_price, a_cr, label = '-1', color = 'red', marker = 'x')
plt.scatter(b_price, b_cr, label = '1', color = 'blue', marker = 'o')
plt.xlabel("Price")
plt.ylabel("Compression Ratio")
plt.legend(loc='upper right')
plt.show()
```
